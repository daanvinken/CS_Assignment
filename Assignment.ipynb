```python
import csv
```

```python
x = []
y = []
z = []
with open('istherecorrelation.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=';')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}\n')
            line_count += 1
        else:
            print(f'{row[0]}  ;;  {row[1]}  ;; {row[2]}.')
            xx = row[1].replace(',','.')
            yy = row[2].replace(',','.')
            zz = row[0].replace(',','.')
                  
            x.append(float(xx))
            y.append(float(yy))
            z.append(float(zz))
            line_count += 1
```

```python
import matplotlib.pyplot as plt
plt.plot(z,x)
plt.plot(z,y)
plt.xlabel
plt.show()
```

```python
# create figure and axis objects with subplots()
fig,ax=plt.subplots()
ax.plot(z,x, marker="o")
ax.set_xlabel("year")
ax.set_ylabel("lifeExp")
ax.plot(z,y, marker="o")
plt.show()
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(z,x, color="red", marker="o")
# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel("WO",color="red",fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(z,y,color="blue",marker="o")
ax2.set_ylabel("NL Beer Consumption [x1000 hectoL]",color="blue",fontsize=14)
plt.show()
# save the plot as a file
fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
